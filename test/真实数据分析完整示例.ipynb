## 1Ô∏è‚É£ Êï∞ÊçÆÂä†ËΩΩ

```
import pandas as pd

data = {
    'year': [2019, 2020, 2021, 2022, 2023, 2024],
    'revenue': [120, 135, 160, 155, 180, 210],
    'cost': [80, 90, 105, 110, 120, 130]
}

df = pd.DataFrame(data)
df
```

## 2Ô∏è‚É£ Êï∞ÊçÆÂ§ÑÁêÜ‰∏éÊåáÊ†áËÆ°ÁÆó

```
df['profit'] = df['revenue'] - df['cost']
df['profit_margin'] = df['profit'] / df['revenue']
df
```

## 3Ô∏è‚É£ Êï∞ÊçÆÂèØËßÜÂåñ

```
import matplotlib.pyplot as plt

plt.plot(df['year'], df['revenue'], label='Revenue')
plt.plot(df['year'], df['cost'], label='Cost')
plt.plot(df['year'], df['profit'], label='Profit')
plt.xlabel('Year')
plt.ylabel('Amount')
plt.title('Company Financial Trend')
plt.legend()
plt.show()
```
